# 통계 2차시: 확률과 분포

## 통계 이론 1회차 복습
통계학을 배우는 이유?<br>
통계란 데이터를 요약하고, 비교하고, 의심하고, 판단하는 방법이다.
> 데이터 분석에서 통계는 선택이 아닌 기본기이다.

데이터의 유형
- 수치형 데이터
    - 이산형: 셀 수 있는 값
    - 연속형: 측정을 통해 얻는 값
- 범주형 데이터
    - 명목형: 순서가 없음
    - 순서형: 순서는 있지만 간격의 의미는 없음
    - 이진형: 2가지의 값만 가진다.

기술통계 <br>
수집한 데이터를 정리하고 요약하고 설명(해석과 표현)하는 데 목적이 있다.
>데이터를 보기 쉽게 정리하고 현 상태를 이해하는 데 집중한다.

- 평균: 
    - 산술평균: 모든 값을 더해 데이터 개수로 나눈 평균
    - 절사평균: 값을 정렬한 뒤 양쪽 끝의 일부 값을 제거하고 계산한 평균
    - 가중평균: 각 값에 중요도를 반영해 계산한 평균

- 중앙값: 데이터를 정렬했을 때 가운데에 위치한 값 (개수가 짝수인 경우 가운데 두 값의 평균을 사용)

- 편차: 각 데이터가 평균에서 얼마나 떨어져 있는지를 나타낸 값 
    - 편차 = 각 값 - 평균
    - 그대로 합하면 +, −가 서로 상쇄


- 분산: 편차의 상쇄 문제를 해결하기 위해 편차를 제곱해서 평균낸 값 
    - 분산 = (편차^2들의 평균)
    - 단위가 제곱이라 해석이 직관적이지 않음


- 표준편차: 분산에 제곱근을 취한 값, 사람이 해석하기 쉽게 만든 변이 지표 
    - "보통 평균에서 이 정도 벗어난다"로 해석
    - 표준편차 = √분산

데이터 분포와 탐색
- 백분위수와 박스플롯
- 도수분포표와 히스토그램
- 밀도추정과 밀도그림, 바이올린 도표
- 이진데이터와 범주 데이터 탐색 (막대플롯)

상관관계(산점도 4~5차시에 함 더함)
- 데이터들끼리 서로 관련이 있는지 확인하는 방법
    - 숫자로는 상관계수, 그림으로는 산점도를 사용한다.

- 상관계수
    - +1에 가까울수록: 함께 증가
    - -1에 가까울수록: 하나가 증가하면 다른 하나는 감소
    - 0에 가까울수록: 뚜렷한 관계 없음

추론통계
- 표본을 이용해 모집단의 특성을 추정하는 통계 방법
> 일부 데이터(표본)를 가지고 전체 데이터(모집단)를 판단한다.

# 확률과 확률변수
확률은 불확실한 세상을 이해하는 언어
>우리는 확률을 통해 불확실성 속에서 더 나은 결정을 내리는 방법을 배울것이다.

통계적 실험<br>
비슷한 사건이 반복될 때 여러가지 가능한 결과를 알 수 있지만 정확히 무슨 결과가 발생할 지는 모를 떄, <br>
이러한 특성을 갖는 문제가 바로 통계학의 연구 및 응용 대상이다.

결과: 실제로 관측된 하나의 경우<br>
사건(event): 표폰 공간의 한 부분집합/그 결과들이 모여있는 집합 (지도의 특정 점)<br>
표본공간: 통계적 실험의 모든 가능한 결과의 집합/일어날수있는 모든 경우의 전체 (지도 전체) {}로 표현<br>
연속형 표본공간: 표본공간의 원소개수가 무한하고 셀수 없는 경우 ??????????
- 측도?

생활속의 비유<br>
구슬 뽑기: 사건 = 빨간색 구슬 뽑기 / 표본공간 = {빨강, 흰색}<br>
가위바위보: 사건 = 가위 / 표본공간 = {가위,바위,보}<br>
고객 행동: 사건 = 구매 / 표본공간 = {구메, 미구매}

사건 == 실제 선택한 하나의 결과<br>
표본공간 == 가능한 모든 결과들의 경우의 수

확률 P(A) <br>
확률: 어떤 사건이 일어날 가능성을 0과 1사이의 실수로 표시하는것<br>
0 = 절대 가능<br>
1 = 절대 불가능<br>
그사이 = 가능성의 크기<br>
사건 a가 일어날 확률 = P(A)<br>

구슬뽑기:<br>
구슬 5개중 빨강색 4개 흰색 1개<br>
p(빵강) = 4/5 = 0.8, P(흰색) = 1/5 = 0.2<br>

>확률은 사건이 일어날 가능성의 크기를 숫자로 표현하는 것이다.

대수의 법칙<br>
시행 횟수가 많아질수록, 실제 결과의 평균은 이론적 기댓값에 가까워진다.

예시
- 동전 앞면 확률 = 0.5
- 10번 던지면 → 앞면 7번 나올 수도 있음
- 10,000번 던지면 → 앞면 비율이 0.5에 매우 가까워짐

> 즉 적게 하면 → 운 영향 큼, 많이 하면 → 확률 이론값에 수렴함


시행 결과 평균 → 표본평균<br>
시행을 무한히 많이 하면 표본평균이 기댓값에 가까워짐<br>

확률변수란?
- 사건을 숫자로 변역해 수학적으로 다룰수 있게 만든것
예: 빨강 4, 흰색 1
x= 

확룰변수는 불확실성을 값으로 다루는 도구로, 표본공간의 각 원소에 하나의 실수값을 대응시키는 함수

확률분포: 숫자가 어떻게 퍼져 있는지 보여주는 지도
- 확률변수로 알아낸 숫자가 어떻게 퍼져있는지 보여주는 지도

이산형 분포: 값이 뚝뚝 끊켜있은 : 막대 그래프
연속형 분포: 값이 무한히 이어짐 : 곡선

보통 분포의 모양을 먼저 봄
분포를 보고 데이터의 특징 안정성 이상치를 판단한다.

기댓값 E(x) = 확률적으로 기대되는 평균값 - 오랫동안 반복했을때 수렴하는 값


- 기대값은 한번의 결과가 아닌 많은 시도(대수의 법칙/표본 크기)의 평균이다.

분산과 표준편차 (매우 중요한 개념이니 계속 나옴 )
- 분산: 평균에서 떨어진 정도(편차)를 제곱 평균
    - 분산 = (편차^2들의 평균)

- 표준편차: 분산에 제곱근을 취한 값, 사람이 해석하기 쉽게 만든 변이 지표 
    - "보통 평균에서 이 정도 벗어난다"로 해석
    - 표준편차 = √분산

- 데이터의 흔들림과 안정성을 보여주는 도구

평균은 == 균형점<br>
표준편차 == 파도의 거칠기

- 분산과 표준편차가 클수록 데이터는 불안정하다
    - 분산과 표준편차는 값들이 얼마나 평균에서 떨어진 정도를 확인하기 때문

# 이론적 확률분포 & 파라미터
사람마다 지문이 다르다, 이는 데이터도 마찬가지로
- 확률분포라는 지문을 보면 집단의 특징이 나타난다.
- 모양을 결정하는 숫자가 존재
- 정규분포는 평균(μ)과 표준편차(σ) 두 숫자로 정의

정리:
모잡단을 N(μ,σ²)로 표현하면, 그 집잔의 성질을 절반은 파악했다 할수 있다.
- N(평균, 분산)


평균μ - 분포의 중심 위치 (대표값): 평균이 올라가면 오른쪽으로 이동
표준편차σ - 분포의 흩어짐 (안정성): 더 넓고 낮아짐 

평균 μ는 분포의 위치를, σ의 모양을 결정한다

확률분포는 데이터의 집단의 지문
파라미터는 그 지문을 읽게 해주는 설명서

# 정규분포란?
종 모양의 대칭 분포로, 대부분의 데이터가 평균 주위에 몰려 있는 분포를 의미
- 평균을 중심으로 좌우 대칭이며, 평균에서 멀어질수록 데이터의 빈도가 감소한다.
- 대부분의 데이터가 평균 주변에 몰려 있으며, 평균에서 멀어질수록 빈도가 줄어든다.

종 모양의 연속함숴
평균에 대해 서로 대항(좌우 대칭), 따라서 평균의 왼쪽, 오른쪽의 확률은 0.5
평균이나 표준편차의 값에 따라 정규분호는 무한히 많이 있음
정규확률변수는 평균 주위에 대부분의 값을 가지고, 평균에서 좌우로 표준편차의 3배 이상 떨어진 값은 없음

평균μ - 분포의 중심 위치 (대표값): 평균이 올라가면 오른쪽으로 이동
표준편차σ - 분포의 흩어짐 (안정성): 더 넓고 낮아짐 

왜하필 종 모양일까?
- 대부분 평균 근처에 몰리기 때문

표준정규분포
평균 = 0 표준편차 = 1 N(0,1)

정리
정규분포는 세상에서 제일 흔한 분포
평균 = 위치
표준편차 = 폭

분석가는 평균과 표준편차를 집단으로 요약

68-95-99 법칙 == 범위를 보면 분포가 보인다.
- 흔한값과 드문값을 빠르게 가르는 직관적인 기준

평균 +-1σ 68
평균 +-2σ 95
평균 +-3σ 99%

표준화(z) - 또 다른 단위도 한 줄로 공정 비교

데이터마다 평균과 표준편차가 다름, 원점수만 비교하면 불공정하다

z = (x-μ)/σ
- 해석: 평균에서 표준편차 몇 배 떨어졌는지 나타내는 숫자

기억하기
주의할점
z는 상대적 지표 희귀성 비교는 가능, 절대적 우열이 아님
정규분포일 때 가장 해석이 직고나적

주의
분포 확인 없이 z만 쓰면 착시 가능

z갑은 평균에서 표준편차에서 

- 평균, 표준편차와 정규분포의 관계
- 정규, 비정규 정규분포는 강력하나 모든 데이터에 항상 맞지 않음 (성격에 따라 구분 가능해야함)
- 정규 근사 (사용 가능)
- 정규 경계 - 오른쪽 꼬리 긴 분포, 해비테이르 이항분포 (주의하며 사용)
- 현실 데이터는 정규보단 꼬리가 긴 경우가 많다

히스토그램
박스플롯
정규에 충분히 근사

정규와 다르면 - 로그정규, 푸아송, 이항, 파레도 대안 분포 고료
