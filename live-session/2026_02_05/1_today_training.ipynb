# 통계 2차시: 확률과 분포

# 통계 이론 1회차 복습
통계학을 배우는 이유?<br>
통계란 데이터를 요약하고, 비교하고, 의심하고, 판단하는 방법이다.
> 데이터 분석에서 통계는 선택이 아닌 기본기이다.

데이터의 유형
- 수치형 데이터
    - 이산형: 셀 수 있는 값
    - 연속형: 측정을 통해 얻는 값
- 범주형 데이터
    - 명목형: 순서가 없음
    - 순서형: 순서는 있지만 간격의 의미는 없음
    - 이진형: 2가지의 값만 가진다.

기술통계 <br>
수집한 데이터를 정리하고 요약하고 설명(해석과 표현)하는 데 목적이 있다.
>데이터를 보기 쉽게 정리하고 현 상태를 이해하는 데 집중한다.

- 평균: 
    - 산술평균: 모든 값을 더해 데이터 개수로 나눈 평균
    - 절사평균: 값을 정렬한 뒤 양쪽 끝의 일부 값을 제거하고 계산한 평균
    - 가중평균: 각 값에 중요도를 반영해 계산한 평균

- 중앙값: 데이터를 정렬했을 때 가운데에 위치한 값 (개수가 짝수인 경우 가운데 두 값의 평균을 사용)

- 편차: 각 데이터가 평균에서 얼마나 떨어져 있는지를 나타낸 값 
    - 편차 = 각 값 - 평균
    - 그대로 합하면 +, −가 서로 상쇄


- 분산: 편차의 상쇄 문제를 해결하기 위해 편차를 제곱해서 평균낸 값 
    - 분산 = (편차^2들의 평균)
    - 단위가 제곱이라 해석이 직관적이지 않음


- 표준편차: 분산에 제곱근을 취한 값, 사람이 해석하기 쉽게 만든 변이 지표 
    - "보통 평균에서 이 정도 벗어난다"로 해석
    - 표준편차 = √분산

데이터 분포와 탐색
- 백분위수와 박스플롯
- 도수분포표와 히스토그램
- 밀도추정과 밀도그림, 바이올린 도표
- 이진데이터와 범주 데이터 탐색 (막대플롯)

상관관계(산점도 4~5차시에 함 더함)
- 데이터들끼리 서로 관련이 있는지 확인하는 방법
    - 숫자로는 상관계수, 그림으로는 산점도를 사용한다.

- 상관계수
    - +1에 가까울수록: 함께 증가
    - -1에 가까울수록: 하나가 증가하면 다른 하나는 감소
    - 0에 가까울수록: 뚜렷한 관계 없음

추론통계
- 표본을 이용해 모집단의 특성을 추정하는 통계 방법
> 일부 데이터(표본)를 가지고 전체 데이터(모집단)를 판단한다.

# 2. 확률: 불확실성을 숫자로 말하기
확률은 불확실한 세상을 이해하는 언어
>우리는 확률을 통해 불확실성 속에서 더 나은 결정을 내리는 방법을 배울것이다.

## 통계적 실험(Experiment)<br>
우연성이 있는 관측/시도 자체를 말한다.
- ex 동전 던지기, 주사위 굴리기, 고객이 구매했는지 관측하기

비슷한 사건이 반복될 때 여러가지 가능한 결과를 알 수 있지만 정확히 무슨 결과가 발생할 지는 모를 떄, <br>
이러한 특성을 갖는 문제가 바로 통계학의 연구 및 응용 대상이다.

### 결과(표본점, Sample Point): 실제로 관측된 하나의 경우
통계 실험에서 실제로 “딱 한 번” 관측된 한 가지 결과를 말한다.

개념설명
- 동전 던지기, 제품 검사, 주문 선택 같은 걸 “통계적 실험”이라고 생각한다.
- 그 실험을 실제로 한 번 실행하면 결과가 하나 나온다. 그 하나가 결과(표본점)이다.

예시
- 가위바위보를 했더니 “가위”가 나왔다면, 이 “가위”가 결과(표본점)이다.

### 표본공간(Sample Space, S): 가능한 모든 결과(표본점)를 모아둔 집합
실험에서 나올 수 있는 “모든 가능한 결과의 목록(집합)”이 표본공간이다.

개념설명
- 표본공간은 “가능한 경우의 수 전체”다. 보통 중괄호 {}로 적는다.
- 분석 관점에서 중요한 포인트는, 먼저 표본공간을 정의해야 “무엇이 가능한지”가 정리된다는 점이다.
- 이산형 표본공간 == 유한 또는 셀 수 있는 무한
    - 공장에서 제품을 반복 생상할때 정삼품 또는 불량품의 확률
- 연속횽 표본공간 == 셀 수 없는 무한
    - 물건이 30분 이내로 배달올 확률

예시설명
- 가위바위보의 표본공간은 {가위, 바위, 보}이다.
- 고객 행동을 단순화하면 {구매, 미구매}처럼 둘로 잡을 수도 있다. 

### 사건(event): 표폰공간의 한 부분집합/그 결과들이 모여있는 집합
사건은 표본공간에서 “관심 있는 결과들만 모아놓은 부분집합”이다. 

개념설명
- 표본공간이 “가능한 전체 지도”라면, 사건은 그 지도에서 내가 관심 있는 구역(부분집합)이다.
- 현실의 질문은 보통 “특정 결과가 나왔나?”가 아니라 “관심 사건이 일어났나?” 형태로 나온다. 
- 예: 구매했나, 클릭했나, 불량이었나\

예시설명
- 구슬 뽑기에서 사건을 “빨간색 구슬 뽑기”로 두고, 표본공간을 {빨강, 흰색}로 둘 수 있다.
- 공장에서 제품이 정상품/불량품으로 나온다면
    - 표본공간: {정상품, 불량품}
    - 사건: “불량품이 나왔다” 또는 “정상품이 나왔다”가 된다.

## 확률이란? — “미래 결과가 얼마나 가능성 있는가”
확률은 사건 A가 일어날 가능성을 0~1 사이의 숫자로 표현한 것

개념설명
- 확률은 불확실한 일을 “숫자”로 다루게 해주는 언어다.
- 0 ≤ P(A) ≤ 1
- 불가능하면 0, 반드시 일어나면 1
- 0과 1 사이 == 사건 A가 일어날 확률 = P(A)
- P(A) = (사건 A에 해당하는 경우 수) / (전체 경우 수) : “사건 A가 일어날 확률”을

예시
- “30대 여성 고객의 쿠폰 클릭 확률이 12%”라면 P(클릭)=0.12로 표현하고, 타깃팅 전략을 설계하는 근거로 쓴다.

### 확률은 예언이 아니다
확률은 “보장”이나 “예언”이 아니라, 장기적으로 기대되는 경향을 숫자로 말하는 것이다.

개념설명
- P(A)=0.7이라고 해서 A가 반드시 일어난다는 뜻이 아니다.
- 단기에는 결과가 흔들릴 수 있고, 확률은 그 흔들림 속에서도 “규칙”을 설명하기 위한 표현이다.

### 대수의 법칙
반복 횟수가 커질수록 관측된 비율(평균)이 실제 확률에 가까워지는 성질이다.

개념설명
- 동전을 10번만 던지면 앞면 비율이 0.2, 0.7처럼 요동칠 수 있다. (표본이 작아 흔들림이 큼)
- 100번, 1000번처럼 많이 반복하면 앞면 비율이 점점 0.5에 가까워진다. <br>
→  이를 대수의 법칙이라 부른다.

예시설명
- 동전 던지기: 10번은 들쭉날쭉, 1000번이면 50%에 수렴한다.

## 확률변수란? — “결과를 숫자로 바꾼 것”
확률변수는 실험 결과(사건)를 숫자 값으로 바꿔주는 규칙(함수)다.


개념설명
- 확률이 "어떤 사건이 일어났는가?"를 다룬다면, 확률변수는 이를 숫자로 표현한다.
- 말 그대로 “랜덤하게 나오는 값”을 하나의 변수 X로 놓는 것이다. 

예시 설명
- 빨강공 또는 흰색공을 꺼낼 확률
- 빨강 4, 흰색 1
    - X= {빨강, 흰색}
    - P(X=빨강) = 0.8
    - P(X=흰색) = 0.2

### 이산형 확률변수(Discrete Random Variable)
이산형 확률변수는 값이 0,1,2,…처럼 “딱딱 떨어지는(셀 수 있는)” 경우

개념설명
- 값들이 연속적으로 이어지는 게 아니라, 가능한 값 목록이 “점”처럼 존재한다.
- 보통 **개수(count)**를 셀 때 많이 사용된다.

예시
- 오늘 고객 문의 건수 X: 0건, 1건, 2건 (정수로 셈)
- 제품 불량 개수 X: 0개, 1개, 2개
- 주사위 눈 X: 1~6 

### 연속형 확률변수(Continuous Random Variable)
연속형 확률변수는 값이 키, 시간처럼 “측정되어 연속적으로 이어지는” 경우

개념설명
- 두 값 사이에 다른 값이 무한히 들어갈 수 있다(예: 170.1과 170.2 사이에도 170.15가 있다).
- 보통 **측정(measurement)**에서 나온다.

예시
- 배송 도착 시간 X: 2.1일, 2.15일, 2.2일…
- 고객 체류시간 X(초): 3.2초, 3.21초…
- 키/몸무게/온도 같은 실수 값들

# 3. 확률분포
각 값이 나올 확률을 표나 함수로 표현

## 확률분포 — “값이 나올 확률 구조”
확률분포는 확률변수 X가 가질 수 있는 값들과, 각 값이 나올 확률을 묶어 정리한 것이다.

개념설명
- 확률변수 X를 정의했으면, 이제 중요한 건 “어떤 값이 얼마나 자주 나오나?”다.
- 확률분포는 그걸 표/그래프/수식으로 정리한 “지도”다.
- 즉, X의 값 영역 전체를 대상으로 확률을 붙여놓은 것이다. 

예시
- 주사위 눈을 X라고 하면, X는 1~6을 가진다.
- “각 눈이 나올 확률이 1/6”이라면, 그 6개 값에 확률이 붙은 것이 확률분포다.

### (이산형) 확률질량함수 PMF:  값의 개수를 셀 수 있다.
이산형에서는 각 값 x에 대해 P(X=x)처럼 ‘딱 그 값이 나올 확률’을 줄 수 있다.


개념설명
- 이산형은 가능한 값이 “점”으로 떨어져 있으니, 각 점에 확률을 바로 붙인다.
- 그래서 P(X=0), P(X=1) 같은 식으로 표현이 자연스럽다. 
- 추가내용: 확률을 각각 더하면 = 1

예시설명<br>
주사위 예시 <br>
P(X=1)=1/6 <br>
P(X=2)=1/6 <br>
P(X=3)=1/6 <br>
P(X=4)=1/6 <br>
P(X=5)=1/6 <br>
P(X=6)=1/6 <br>
→ 모든 확률 합 = 1

### 베르누이 분포 (Bernoulli Distribution) — 이산형(PMF)
결과가 성공/실패 두 가지뿐인 실험에서, X가 0 또는 1만 가지는 분포다.

개념설명
- 베르누이 시행: 가능한 결과가 두 가지인 반복 실험(불량/양호, 찬성/반대).
- 베르누이 확률변수 X는 보통
    - 성공이면 X=1, 실패면 X=0
- 분포는 PMF로 표현한다(이산형이라 P(X=1), P(X=0) 형태).

예시설명
- 클릭 여부 X: 클릭=1, 미클릭=0
- P(X=1)=p, P(X=0)=1-p (합은 1)

### 이항분포 (Binomial Distribution) — 이산형(PMF)
성공확률 p가 동일한 베르누이 시행을 독립적으로 n번 반복했을 때, X=성공 횟수의 분포다.

개념설명
- 조건(핵심만):
    - 결과가 성공/실패 두 가지
    - 성공확률 p가 매번 동일
    - 각 시행이 독립적
    - 시행 횟수 n이 고정
- 표기: B(n, p) (n과 p가 모수).
- X는 성공 횟수라서 0,1,2,…,n만 가능 → 이산형.

예시설명
- 20개 제품 검사에서 불량(성공)을 몇 개 찾는가
    - X=불량 개수 → B(20,p)

### 포아송분포 (Poisson Distribution) — 이산형(PMF)
일정 시간/공간에서 희귀 사건이 몇 번 발생하는지(횟수)를 나타내는 분포다.

개념설명
- “아주 작은 확률로 발생하는 사건(희귀)”의 발생 횟수에 적용. 
- 독립성: 서로 다른 구간의 발생 횟수는 독립
- 비집략성: 아주 작은 구간에서 2번 이상 동시에 일어날 확률은 0에 가깝다
- 비례성 : 단위 시간/공간에서 평균 발생 횟수는 일정
- X는 0,1,2,… (횟수) → 이산형.

예시설명
- 하루 정전 횟수
- 화재보험, 위조수표 같은 희귀 사건 횟수


### (연속형) 확률밀도함수 PDF, 면적=확률
연속형에서는 P(X=x)=0이고, 구간 확률은 ‘곡선 아래 면적’으로 계산한다. 

개념설명
- 연속형은 값이 무한히 촘촘하게 존재한다. 그래서 “정확히 x가 나올 확률”을 따지면 0이 된다
- 대신 “a와 b 사이에 있을 확률”을 계산한다.
- 이때 확률은 그래프(밀도곡선) 아래 면적으로 해석한다. 

예시설명
- 배송시간 X가 연속형일 때
    - “정확히 2.000일 걸릴 확률”은 0
    - 대신 “1.8일~2.2일 사이일 확률”처럼 구간으로 말한다.

### 정규분포 (Normal Distribution) — 연속형(PDF)
평균 μ를 중심으로 좌우 대칭인 종 모양 연속 분포로, 퍼짐은 σ가 결정한다.

개념설명
- 연속형이므로 “정확히 X=어떤 값”의 확률은 0이고, 구간 확률(면적)로 계산한다
- 종 모양의 연속함수 
- 평균 μ 기준 좌우 대칭 → 왼쪽/오른쪽 확률 0.5씩
- μ, σ 값에 따라 정규분포는 무한히 많다


### 확률분포의 기본 성질
확률은 전부 더하면 1이고, 연속형은 전체 구간 면적이 1이다. 

- 이산형: 모든 가능한 값에 대한 확률을 더하면 1이 된다.
- 연속형: 전체 구간에서 곡선 아래 면적이 1이 된다.
- 이는 확률의 총량이 1 이다 라는 말과 같다.

### (분포 요약) 기댓값과 분산은 “분포의 대표값/퍼짐”이다

#### 기댓값 E(X)
기댓값은 확률을 고려한 평균, 즉 장기적으로 기대되는 중심값이다

개념설명
- “평균”을 구하되, 각 값이 나올 확률을 가중치로 둔다. 그래서 분포의 중심을 대표한다.

예시설명
- 게임에서 점수 X가 {0점(확률 0.5), 10점(확률 0.5)}라면
- 기대되는 평균 점수는 5점이다.

#### 분산 Var(X), 표준편차 SD(X)
분산/표준편차는 값들이 기댓값 주변에서 얼마나 퍼져 있는지를 나타낸다

개념설명
- 평균이 같아도 분포 모양은 다를 수 있다.
- 그래서 퍼짐(흩어짐)을 숫자로 표현해야 하고, 그게 분산/표준편차다

예시설명
- A반: 점수가 48~52에 몰려있음(평균 50, 퍼짐 작음)
- B반: 점수가 10~90까지 넓게 퍼짐(평균 50, 퍼짐 큼)
- 평균만 보면 같지만 “안정성/변동성”은 완전히 다르다.

# 4. 정규분포

## 정규분포란?: “현실 데이터의 기본 모양”
정규분포는 가운데(평균) 근처에 값이 가장 많고, 좌우 대칭인 종 모양 분포다.

개념설명
- 많은 현실 데이터가 “대충 종 모양”에 가까운 경우가 많아서 자주 쓰는 기준 분포다.
- 정규분포를 쓰는 이유는 “확률 계산”을 표준화/표준정규표로 쉽게 할 수 있기 때문이다.

예시설명
- 사람 키, 시험 점수, 측정 오차 등은 정규분포로 근사해서 다루는 경우가 많다(단, 항상 그런 건 아님). 

### 정규분포 결정요소: 평균 μ, 표준편차 σ
정규분포는 평균 μ가 위치를, 표준편차 σ가 퍼짐(폭)을 결정한다.

개념설명
- μ가 바뀌면 그래프 전체가 좌우로 이동한다(중심 이동). 
- σ가 커지면 더 넓고 낮게 퍼지고, σ가 작으면 더 좁고 높게 몰린다(폭 변화).

예시설명
- A반 키가 평균 170cm, B반 평균 175cm라면(σ가 비슷하다고 가정) B반 분포가 오른쪽으로 이동한 형태다.
- 같은 평균 170cm라도 σ가 큰 반은 키 차이가 크고, σ가 작은 반은 키가 비슷비슷하다.

### 68–95–99 법칙(경험법칙)
정규분포에서는 평균 주변 ±1σ, ±2σ, ±3σ 안에 들어오는 비율이 대략 68%, 95%, 99%다.

개념설명
- 정규분포는 “평균에서 얼마나 떨어졌는지”로 확률을 빠르게 감 잡을 수 있다.

예시설명
- 시험 점수가 정규분포이고 평균 70, 표준편차 10이라면
    - 60~80점(±1σ)에 대략 68%
    - 50~90점(±2σ)에 대략 95%가 들어간다.


## 정규분포를 쓰기 전 체크
정규분포는 자주 쓰지만, 데이터가 정규에 가까운지(모양) 확인하고 쓰는 게 안전하다.

개념설명
- 현실 데이터는 비대칭이거나 꼬리가 두꺼운 분포도 많다.
- 그때 정규분포를 무리하게 적용하면 확률/구간 해석이 틀어질 수 있다.

예시설명
- 소득, 매출 같은 데이터는 한쪽으로 긴 꼬리를 가진 경우가 많아서 정규가 아닐 수 있다

# 5. 표준화(Z-score)와 표준정규분포

## 표준화(Standardization)란?
표준화는 임의의 정규분포 값을 공통 기준 N(0,1) 눈금으로 바꾸는 변환이다.

개념설명
- 서로 평균/표준편차가 다른 정규분포끼리는 값 크기를 직접 비교하기 어렵다.
- 그래서 “평균에서 몇 표준편차 떨어졌는지”로 바꿔 비교/확률계산을 한다.

예시설명
- A시험(평균 70, 표준편차 10)에서 85점, B시험(평균 80, 표준편차 5)에서 88점
- 그대로 비교하면 애매하지만, Z로 바꾸면 어느 쪽이 상대적으로 더 높은지 비교가 된다.

## 표준화 작업

1. 문제의 X를 확인하고, 정규분포의 μ, σ를 정리한다.
2. z = (X−μ)/σ로 표준화한다. == (값- 평균) / 표준편차
