# 모집단과 표본이란

## 모집단과 표본
- 모집단: 관심의 대상이 되는 전체 집단 
    - 예시: 한 국가의 모든 성인, 특정 음식의 전체 판매량 등
    - 매우 크기 때문에 분석에서 제대로 활용하기 어려움
    - 다루기 힘듦

- 표본: 모집단에서 추출한 일부 집단
    - 예시: 그 국가의 성인 중 일부, 특정 음식의 지역 판매량 등

### 표본을 왜 사용할까?
1. 현실적 제약: 비용과 시간 그리고 접근성 <br>
모집단 전체를 조사하는 전수조사는 막대한 비용과 시간이 소요되며, 대부분의 경우 현실적으로 불가능하거나 비효율적이다. <br>
표본 조사는 이러한 제약을 극복하면서도 유의미한 분석 결과를 얻기 위한 현실적인 대안이다.

2. 대표성 <br>
잘 설계된 표본은 모집단의 특성을 충분히 반영할 수 있다. 따라서 표본에서 얻은 결과를 모집단 전체로 일반화할 수 있다. <br>
특히 무작위 추출을 사용하면 편향을 최소화하고 모집단의 다양한 특성을 포함할 가능성이 높아진다.

3. 데이터 관리
표본 데이터를 사용하면 전체 데이터를 다루는 것보다 처리와 분석이 훨씬 수월하다.<br>
계산 비용 감소, 컴퓨팅 자원 부담 감소, 오류나 이상치 탐지 및 수정이 용이

4. 모델 검증에 용이
표본 데이터를 이용해 통계적 모델을 학습하고 검증할 수 있다. <br>
모델이 대표성 있는 표본에서 잘 작동한다면, 모집단에서도 잘 작동할 가능성이 높다.

## 조사방법
- 전수조사 <br>
모집단 전체를 조사하는 방법은 규모가 커질수록 시간과 비용이 많으 소비된다.

- 표본 조사 <br>
 표본만 조사하는 방법. 비용과 시간이 적게 드나 표본이 대표성을 지녀야 한다.

## 실제 사용 예시

````python
import numpy as np
import matplotlib.pyplot as plt

# 모집단 생성 (예: 국가의 모든 성인의 키 데이터)
population = np.random.normal(170, 10, 1000)

# 표본 추출
sample = np.random.choice(population, 100)

plt.hist(population, bins=50, alpha=0.5, label='population', color='blue')
plt.hist(sample, bins=50, alpha=0.5, label='sample', color='red')
plt.legend()
plt.title('population and sample distribution')
plt.show()
````

# 표본 오차와 신뢰구간

## 표본오차란?
표본에서 계산된 통계량과 모집단의 진짜 값 간의 차이
- 표본이 클수록 표본 오차는 작아짐 (데이터가 많아짐을 의미)
- 표본이 모집단을 완벽하게 대표하지 못하기 때문에 발생
    - 표본의 크기와 추출 방법에 따라 달라짐
    - 크기: 크기가 클수록 오차는 줄어듬: 더 많은 데이터를 수집할수록 모집단을 더 잘 대표하게 됨
    - 추출 방법:  무작위 추출 방법을 사용하면 표본오차를 줄일 수 있다.
        - 모든 모집단 요소가 선택될 동등한 기회를 가지게 해야한다.

## 신뢰구간
신뢰구간은 모집단의 특정 파라미터(예: 평균, 비율)에 대해 추정된 값이 포함될 것으로 기대되는 범위를 나타낸다.
- 신괴구간 = 표본평균±z×표준오차
    - 여기서 z는 선택된 신뢰수준에 해당하는 z-값
        - 예를 들어, 95% 신뢰수준의 z-값은 1.96

# 정규분포

## 정규분포란?
종 모양의 대칭 분포로, 대부분의 데이터가 평균 주위에 몰려 있는 분포를 의미
- 평균을 중심으로 좌우 대칭이며, 평균에서 멀어질수록 데이터의 빈도가 감소한다.
- 대부분의 데이터가 평균 주변에 몰려 있으며, 평균에서 멀어질수록 빈도가 줄어든다.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 정규분포 생성
normal_dist = np.random.normal(170, 10, 1000)

# 히스토그램으로 시각화
plt.hist(normal_dist, bins=30, density=True, alpha=0.6, color='g')

# 정규분포 곡선 추가
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 170, 10)
plt.plot(x, p, 'k', linewidth=2)

plt.title('normal distribution histogram')
plt.show()
````

# 긴 꼬리 분포

## 긴 꼬리 분포란?
대부분의 데이터가 분포의 한쪽 끝에 몰려 있고, 반대쪽으로 긴 꼬리가 이어지는 형태의 분포
- 정규분포와 달리 대칭적이지 않고 비대칭적
- 특정한 하나의 분포를 의미하지 않으며 여러 종류의 분포를 포함한다
    - 레토 분포, 지프의 법칙, 멱함수
- 표본이 증가해도 정규분포가 될수 없음
- 소득 분포, 웹사이트 방문자 수 등에서 관찰됨.

````python
# 긴 꼬리 분포 생성 (예: 소득 데이터)
long_tail = np.random.exponential(1, 1000)

# 히스토그램으로 시각화
plt.hist(long_tail, bins=30, density=True, alpha=0.6, color='b')
plt.title('long tail distribution histogram')
plt.show()
````

# 스튜던트 T 분포

## 스튜던트 T 분포란?
표본이 적어 평균을 잘못 추정할 위험이 클 때, 그 불확실함을 반영한 분포

### 왜 사용할까?
보통 평균을 볼 때는 정규분포를 사용한다. <br>
정규분포는 모집단의 표준편차를 알고 있고, 데이터가 많을 때 잘 맞는다. <br>
하지만 현실에선 
- 모집단의 표준편차를 모를때가 많다
- 표본이 작다.

이러한 요소 때문에 평균추청은 매우 불안정해진다. → 이 불안정을 반영해서 꼬리를 두껍게 만든 게 t분포다.

### 자유도와 꼬리가 두껍다는 말의 의미
- 표본이 적다 → 평균이 흔들릴 가능성 큼
    - 극단적인 값이 나올 가능성도 고려해야 함
    - 꼬리를 두껍게

- 표본이 많아진다 → 평균이 안정됨
    - 굳이 조심할 필요 없음
    - 정규분포에 가까워짐

### 어떨때 사용할까?
표본이(데이터가) 적은 경우 사용한다.  


````python
# 스튜던트 t 분포 생성
t_dist = np.random.standard_t(df=10, size=1000)

# 히스토그램으로 시각화
plt.hist(t_dist, bins=30, density=True, alpha=0.6, color='r')

# 스튜던트 t 분포 곡선 추가
x = np.linspace(-4, 4, 100)
p = stats.t.pdf(x, df=10)
plt.plot(x, p, 'k', linewidth=2)
plt.title('student t distribution histogram')
plt.show()
````

# 카이제곱 분포

## 카이제곱 분포란?
범주형 데이터의 독립성 검정이나 적합도 검정에 사용되는 분포
- 자유도에 따라 모양이 달라짐.
- 상관관계나 인과관계를 판별하고자 하는 원인의 독립변수가 ‘완벽하게 서로 다른 질적 자료’일 때 활용
    - ex) 성별이나 나이에 따른 선거 후보 지지율
    - 범주형 데이터 분석에 사용

````python
# 카이제곱분포 생성
chi2_dist = np.random.chisquare(df=2, size=1000)

# 히스토그램으로 시각화
plt.hist(chi2_dist, bins=30, density=True, alpha=0.6, color='m')

# 카이제곱분포 곡선 추가
x = np.linspace(0, 10, 100)
p = stats.chi2.pdf(x, df=2)
plt.plot(x, p, 'k', linewidth=2)
plt.title('chi-square distribution histogram')
plt.show()
````

# 이항 분포

## 이항분포란?
같은 실험을 여러 번 했을 때, 성공이 몇 번 나오는지를 나타내는 분포
    - 이산형 분포
        - 연속적으로 이어진 곡선이 아니라, 점으로 찍히는 분포

### 언제 사용하나?
- 시행을 여러 번 반복한다 (n번)
- 매번 결과는 두 가지뿐이다 (성공 / 실패)
- 각 시행은 서로 영향 없다 (독립)
- 성공 확률 p는 항상 같다

### 간단정리
- 이항분포 == 성공 횟수 분포
- 결과는 정수만 나옴 → 이산형 분포
- 기준은 n(횟수), p(성공확률)

````python
# 이항분포 생성 (예: 동전 던지기 10번 중 앞면이 나오는 횟수)
binom_dist = np.random.binomial(n=10, p=0.5, size=1000)

# 히스토그램으로 시각화
plt.hist(binom_dist, bins=10, density=True, alpha=0.6, color='y')
plt.title('이항 분포 히스토그램')
plt.show()
````

# 푸아송 분포
단위 시간 또는 단위 면적 당 발생하는 사건의 수를 모델링할 때 사용하는 분포


## 푸아송 분포란?
단위 시간 또는 단위 면적 당 발생하는 사건의 수를 모델링할 때 사용하는 분포
- 평균 발생률 λ를 가진 사건이 주어진 시간 또는 공간 내에서 몇 번 발생하는지를 나타낸다.
    - 단위 시간 또는 단위 면적당 희귀하게 발생하는 사건의 수를 모델링하는 데 적합

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 푸아송 분포 파라미터 설정
lambda_value = 4  # 평균 발생률
x = np.arange(0, 15)  # 사건 발생 횟수 범위

# 푸아송 분포 확률 질량 함수 계산
poisson_pmf = poisson.pmf(x, lambda_value)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(x, poisson_pmf, alpha=0.6, color='b', label=f'Poisson PMF (lambda={lambda_value})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.legend()
plt.grid(True)
plt.show()
````

# 분포 정리

## 지금까지 본 분포들은 극히 일부에 불과
하지만 결국 데이터 수가 엄청 많아지면 정규분포에 수렴하게 된다. (중심극한정리)
- 데이터 수가 많으면 정규분포로 가정
- 데이터 수가 적으면 각 상황에 맞는 분포 선택
    - 특히, long tail distribution은 데이터가 많아도 정규분포가 되지 않는 분포!

## 어떻게 고르는게 좋을까?
- 데이터 수가 충분하다 → (무조건) 정규분포에 근사
- 데이터 수가 작다 → 스튜던트 t 분포
- 일부 데이터가 전체적으로 큰 영향을 미친다 → 롱 테일 분포 (파레토 분포)
- 범주형 데이터의 독립성 검정이나 적합도 검정 → 카이 제곱 분포
- 결과가 두 개(성공 or 실패)만 나오는 상황 → 이항 분포
- 특정 시간, 공간에서 발생하는 사건 → 푸아송 분포

### 2. 어떤 표본의 평균이 50이고, 표본 표준편차가 10인 경우, 95% 신뢰구간을 구하세요. (표본 크기는 충분히 크다고 가정)
```
95% 신뢰구간은 표본 평균 ± (1.96 * 표본 표준오차)로 계산됩니다.
표본 표준오차는 표준편차를 표본 크기의 제곱근으로 나눈 값입니다.
```

````python
sam_avg = 50 # 표본 평균
sam_std = 10 # 표본 표준편차
n = 100 # 표준 크기
se = sam_std / (n ** 0.5)
lower = sam_avg - 1.96 * se
upper = sam_avg + 1.96 * se

lower, upper

# 95%의 확률로 모집단의 평균은 위의 신뢰구간 범위 안의 값을 가진다.
````

### 7. 카이제곱분포(Chi-Square Distribution)는 주로 어떤 분석에 사용되나요?
1) 두 변수 간의 상관관계 분석
2) 평균 간의 차이 분석
3) 범주형 데이터의 독립성 검정
4) 연속형 데이터의 분포 분석

3번

- 카이제곱분포는 주로 범주형 데이터의 독립성 검정에 사용된다.

### 8. 동전 던지기를 10번 했을 때, 앞면이 6번 나올 확률을 구할 때 사용하는 분포는 무엇인가요?
1) 정규분포
2) 이항분포
3) 푸아송분포
