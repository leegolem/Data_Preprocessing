# 통계 3차시: 추론통계

# 1. 통계적 추론: 추론 통계

잠깐 짚고 넘어가기
- 기술통계: 데이터의 상태를 숫자(평균, 중앙값, 분산, 그래프 등)로 요약해서 빠르게 파악하는 통계
- 추론통계: 표본으로 모집단(전체)의 특성(모수)을 추정하거나, 주장/가설을 검정하는 통계다. 

## 추론 통계의 출발점: "숫자(점) 하나 말고, 구간으로 말한다"
왜 구간으로 말할까?

- "내일 25도일 거다" -> 틀릴 가능성이 매우 높은 예측 (하나의 값으로 예측하기 떄문)
- "내일 온도는 23~27도일 거다" 맞을 가능성이 있는 예측 (구간안에 있는 값이 나올수도 있기 때문)

> 추론 통계는 불확실성을 숫자(구간)로 표현하는 방법이다.

# 2. 예언구간 vs 신뢰구간

### 예언구간(Prediction Interval) = 개별 미래 값
다음에 새로 관측될 1개의 값이 들어갈 법한 범위를 잡는 구간이다.

- 개념설명<br>
예언구간은 다음 관측 1개(내일 기온 1번, 다음 제품 1개, 다음 사람 1명)처럼 개별 결과를 대상으로 한다.
때문에 평균(모수)보다 흔들림이 커서, 보통 범위가 더 넓게 느껴진다.

- 예시
    - 내일 기온은 25도 대신
    - 내일 기온은 23~27도처럼 **다음 하루(개별 값)**를 범위로 말한다.

정규분포 N(μ, σ)를 안다고 가정했을 때

1. 신뢰수준 95%를 고른다. → 표준정규에서 95%가 대략 −1.96 ~ +1.96
2. 원래 단위로 되돌린다. → N(μ, σ)라면 (μ − 1.96σ) ~ (μ + 1.96σ)
3. 예시 <br>
μ=50, σ≈5이면<br>
50 ± 1.96×5 = 40.2~59.8(대략 40~60)

### 신뢰구간(Confidence Interval)
표본으로 계산한 평균을 바탕으로 모집단 평균 μ(모수)가 있을 법한 범위를 추정하는 구간이다.

- 개념설명<br>
신뢰구간은 "개별 1명 결과"가 아닌, 모집단의 진짜 평균(μ) 같은 **고정된 값(모수)**를 겨냥한다. <br>
    1. 표본을 뽑아 평균 x̄를 만든다. <br>
    2. x̄는 표본을 다시 뽑으면 조금씩 달라진다(흔들린다)<br>
    3. 그 흔들림을 반영해서 "μ가 있을 범위"를 만든다

- 예시
    - 한국 성인 남성 평균 키 μ는 전수조사가 어렵다. 그래서 표본 평균을 이용해
    - "μ는 167~173cm 사이일 것 같다"처럼 범위로 말하는 게 신뢰구간이다. (예시 구조).

표준오차 SE
- SE = σ/√n
- n(표본 수)이 커지면 √n이 커져서 평균의 흔들림이 줄어든다(그래서 구간이 좁아짐). 

σ를 아는(또는 가정하는) 단순 버전: z 사용
- 95%면 표준정규의 ±1.96을 사용한다
- x̄±1.96×(σ/√n) # 이때 (σ/√n)은 SE다

σ를 모르는 버전: s와 t 사용
- 예시 데이터가 주어졌을 때<br> (μ=170, n=10, x̄=167.9, s=6.9),95% 신뢰구간을
- 167.9 ± 2.26 × (6.9/√10) ≈ 162.97 ~ 172.83 로 계산한다

- z 사용과 t 사용의 차이점
    - σ(모집단의 표준편차)를 모르면 s를 쓰고 t를 쓴다는 흐름이 나온다.


- σ: 모집단의 표준편차
- s: 표본의 표준편차
- μ: 모집단의 평균
- x̄̄: 표본의 평균
- n: 표본 개수(표본 크기)
- z: z분포(표준정규)에서 가져온 임계값 (95%면 1.96)
    - 95% 양쪽 구간이면 1.96
- t: t분포에서 가져온 임계값 
    - σ를 모르고 s를 쓰는 상황에서 사용 (σ 모를 때, df=n−1에 따라 달라짐)


### 예언구간 vs 신뢰구간
공통점
- 표본으로 계산한 값은 흔들린다. 그래서 불확실성을 구간으로 표현한다.

차이점
- 구간이 겨냥하는 대상이 다르다 
    - 예언구간: 미래의 새 관측값 1개 : "다음 한 명의 점수는 어디쯤?"
    - 신뢰구간: 모집단의 모수(대표로 μ) : "전체 반의 평균 점수는 어디쯤?"

- 해석의 차이
    - 예언구간 해석: "다음 관측값이 이 구간 안에 들어올 것이다(확률적)."
    - 신뢰구간 해석: "같은 방법으로 구간을 많이 만들면, 그중 약 95%가 μ 를 포함한다"


이렇게도 보기
- 예언구간: "다음 1개 값이 어디쯤?" → N(μ,σ) 가정 시 μ ± 1.96σ
- 신뢰구간: "진짜 평균 μ가 어디쯤?" → x̄ ± (임계값) × (SE), SE=σ/√n, σ 모르면 t 사용

판단법

- 다음에 새로 나올 관측값 1개가, 높은 확률로 이 구간 안에 들어올 것이다!→ 예언구간
    - 다음 1개의 값 범위
- 표본평균으로 만든 이 구간은, 신뢰수준(예: 95%)으로 모평균 μ를 포함하도록 만든 구간이다! → 신뢰구간
    - 모평균 μ 범위

# 3. 95% 신뢰구간과 표준오차

## 표준정규에서 95%: "±1.96" (임계값 z)
95%를 범위로 잡을 때, 표준정규에서는 중앙에서 ±1.96만큼이 기준이 된다.

개념설명
- "95% 신뢰수준"은 중앙에 95%를 담겠다는 뜻이다.
- 표준정규(N(0,1))에서는 중앙 95%가 대략 -1.96 ~ +1.96 구간이다.
- 이 ±1.96은 "95%를 담기 위한 기준선"으로 **임계값(critical value)**이라고 부른다.

예시설명
- 표준정규에서 어떤 값이 z=2.5라면? 중앙 95% 바깥쪽에 있을 가능성이 크다 <br>
왜냐면 1.96보다 크니까
- 반대로 z=1.2면 중앙 95% 안쪽에 있을 가능성이 크다.<br>
왜냐면 1.96보다 작으니까

## 표준오차 SE: "표본평균이 흔들리는 정도"
SE는 표본평균이 얼마나 흔들리는지를 숫자로 나타낸 것이다.

개념설명
- 표본을 다시 뽑으면 표본평균 x̄는 매번 조금씩 달라진다.
- 이 흔들림의 크기를 나타내는게 표준오차(SE)다.
- σ를 안다고 가정할 경우 SE = σ/√n
    - n이 커질수록 √n이 커진다 → σ/√n이 작아진다 (SE가 감소)
    - 즉 평균이 덜 흔들리고, 신뢰구간 폭이 줄어든다.

예시설명
- σ=10일 경우 (모집단 표준편차가 10)
    - n=25면 SE = 10/√25 = 10/5 = 2
    - n=100이면 SE = 10/√100 = 10/10 = 1<br>
    → 표본을 더 많이 모으면 평균이 더 안정적이라 SE가 줄어든다.

## 95% 신뢰구간(기본형): “x̄ ± 1.96×SE” (σ를 아는 경우)
95% 신뢰구간은 표본평균 x̄를 중심으로, ±(1.96*SE)만큼의 범위를 만든 것이다.

개념설명
- 신뢰구간의 뼈대는 이 구조다: 신뢰구간 = 중심값(추정값) ± 오차범위
    - 중심값: 표본평균x
    - 오차범위: (임계값)*(표준오차 SE)
- σ를 안다고 가정할 경우:
    - SE = σ/√n
    - 임계값(95%) = 1.96
    - x̄ ± 1.96×(σ/√n) 가 된다. 

예시설명
- x̄=50, σ=10, n=100이면
    - SE=10/√100=1
    - 95% 신뢰구간 = 50 ± 1.96×1 = 48.04 ~ 51.96<br>
    → 표본평균 50 근처에 “모평균 μ가 있을 범위”를 만든다.

## 95% 신뢰구간 해석
95% 신뢰구간은 “이번 구간이 맞을 확률”이 아니라 구간을 만드는 방법의 장기적 성공률이다

개념설명
- μ가 95% 확률로 이 구간 안에 있다: 틀린 해석
- 같은 과정을 100번 반복해 구간을 만들면 그 중 약 95개의 구간이 μ를 포함한다: 맞는 해석

다시 한번 보기
- 표본을 다시 뽑을 때마다 x̄가 달라지고, 구간도 달라진다.
- μ는 고정된 값, 구간은 매번 바뀐다.
- 95%는 그 바뀌는 구간들이 μ를 잡아내는 비율이다.

## σ를 모르는 현실 버전: s와 t를 사용
모표준편차 σ를 모르니까 표본표준편차 s로 대신하고, 이때는 1.96(z) 대신 t값을 쓴다.

개념설명
- 대부분 현실에선 모집단 표준편차 σ를 모른다. (비용/시간/물리적으로 힘듦)
- 그래서 표본으로 계산한 표준편차 s를 사용
- σ 대신 s를 쓰면 불확실성이 더 커지기 쉬워 95% 구간을 만들 때 z(1.96) 대신 t 분포의 임계값으로 사용

예시설명
- n=10, x̄=167.9, s=6.9, 95%에서 t=2.26
    - 95% 신뢰구간 = 167.9 ± 2.26×(6.9/√10) ≈ 162.97 ~ 172.83

# 4. 신뢰구간을 “좋게” 만드는 핵심 개념들

## 표본크기 n과 √n 효과
n(표본 수)이 커질수록 신뢰구간은 좁아져서 더 정확해진다(폭이 √n에 의해 줄어듦).

개념설명
- 신뢰구간 폭은 표준오차(SE)에 의해 결정되고, SE는 대략 (σ or s)/√n 꼴이다.
- 그래서 n이 커지면 √n이 커지고 → SE가 줄어들고 → 구간이 좁아진다.

예시설명
- 표준편차가 비슷하다고 가정할 때
    - n=25 → √n=5
    - n=100 → √n=10 (표본이 4배)
    - √n이 2배가 되니까, SE(폭)는 절반 수준으로 줄어든다

## Bias(편향): n을 늘려도 해결 안 되는 문제
표본이 치우치게 뽑히면(Bias), n을 크게 해도 추정이 틀어진다.

개념설명
- 신뢰구간은 “표본이 모집단을 잘 대표한다”는 전제 위에서 의미가 있다.
- 만약 표본을 뽑는 과정이 한쪽으로 치우치면(편향), 평균도 한쪽으로 치우친다.
- 이 경우 표본을 아무리 많이 모아도 “정확해지는 방향”이 아닌 “치우친 값이 더 확신 있게” 나올 수 있다.

개념설명
- 한국 성인 남성 근육량을 추정하려는데 표본을 헬스장 이용자만 뽑으면?<br>
→ 평균이 실제보다 커질 가능성이 크다(표본이 대표성을 가지지 않음).<br>
→ 이 상태에서 n만 늘리면 “헬스장 이용자 평균”을 더 정밀하게 만들 뿐, “전국 평균”에 가까워지진 않는다.

## CLT(중심극한정리): 왜 신뢰구간이 가능한가(근거)
n이 충분히 크면 표본평균 x̄의 분포가 정규분포에 가까워져서, z(또는 t)로 구간을 만들 수 있다.

개념설명
- 신뢰구간은 결국 “표본평균이 얼마나 흔들리는지”를 정규분포(또는 t) 성질로 계산해서 만든다.
- 그런데 원래 데이터가 꼭 정규분포일 필요는 없더라도, 표본평균 자체는 n이 커지면 정규에 가까워진다는 정리가 CLT다.
- 그래서 신뢰구간에서 ±1.96 같은 임계값을 쓰는 논리가 성립한다

예시설명
- 한 사람의 구매금액은 분포가 찌그러져 있을 수 있다(비정규).
- 하지만 “50명씩 뽑아서 평균 구매금액”을 여러 번 만들면, 그 평균들은 정규처럼 모이는 경향이 생긴다. <br>
→ 그래서 평균에 대한 신뢰구간이 가능해진다.

## t와 근사(연결): 왜 어떤 경우는 z 대신 t를 쓰나
σ를 모르고 s로 대체하면 불확실성이 커져서, z 대신 t 임계값으로 보정한다. <br>
(그리고 n이 커지면 t는 z에 가까워진다)

개념설명
- 현실에서는 모집단 표준편차 σ를 모르는 경우가 많아 표본표준편차 s를 쓴다.
- 이때 신뢰 구간을 만들 때 z(1.96) 대신 t 임계값을 사용한다.
- 표본이 작고(n 작음), σ도 모르면 → 불확실성 ↑ → t 사용(보통 더 넓어짐)
- 표본이 커지면 → t가 z에 가까워진다(근사)

에시설명
- n=10처럼 작으면 t 임계값이 1.96보다 큰 경우가 많아서(예: 2.xx) 구간이 더 넓어진다.
- n이 커질수록 t 임계값은 1.96 근처로 내려오는 방향("z에 가까워짐")이다.

# 1. 가설검정
표본을 뽑으면 값이 흔들린다(표본오차). <br>
그래서 “관측된 차이/효과가 **우연(표본변동)**으로도 충분히 나올 만한가?”를 따져야 한다

가설검정 (살짝 찍고 가볍게 듣기)
차이가 우연일까 진짜일까?

## 가설검정이란?
표본에서 보이는 차이가 우연인지(샘플링 변동), 진짜 효과인지 판단하는 절차다.

개념설명
- 모집단으로 부터 표본을 뽑으면 결과는 매번 달라진다.
- 때문에 지금 관측된 차이가 그냥 우연에 의한것인지 충분히 따져봐야 한다.
- 가설검정은 이걸 규칙(p-value, α)으로 결정한다.

예시설명
- A/B 테스트에서 A 평균 52,000원, B 평균 50,000원이라면<br>
“2,000원 차이가 진짜로 의미 있는 차이냐?”를 판단한다.

### 귀무가설 H0
차이/효과 없음을 기본값으로 두는 가설

개념설명
- 가설검정은 일단 "변화 없다"를 출발점으로 둔다.
- 그리고 데이터가 그 가정을 얼마나 흔드는지 확인한다.
- H0는 보통 “같다”, “0이다”, “차이가 없다” 형태다.

예시
- H0: “A와 B의 평균은 같다(차이 = 0)” 

### 대립가설 H1
차이/효과 있음을 주장하는 가설: 보통 우리가 얻고자 하는 결과

개념설명
- 우리가 궁금한 건 보통 H1이다(효과가 있나?).
- 다만 검정은 H1을 직접 “증명”하기보다,
- H0가 말이 안 될 정도로 희박하면 H0를 버리는 방식이다.

예시
- H1: “A와 B의 평균은 다르다(차이 ≠ 0)“

## p-value?
H0가 참이라고 가정했을 때, 지금 같은 결과(또는 더 극단)가 나올 확률이다.

개념설명
- p-value는 “H0가 맞다”는 전제 아래에서 계산한다.
- p가 작을수록: “H0가 맞다면 이런 결과는 거의 안 나옴” → H0가 의심됨
- p가 클수록: “H0여도 이런 결과는 나올 수 있음” → H0를 버릴 근거가 약함

예시설명
- 동전을 10번 던졌는데 10번 연속 앞면이 나왔다고 하자.
- 공정하다(H0), 불공정하다(H1)
    - H0: “공정한 동전(앞면 확률 0.5)”라면 그 확률은 (1/2)^10 = 0.00098로 매우 작다.  <br>
    → “공정하다(H0)” 가정이 좀 이상해 보인다.


## 유의수준 α(알파)
H0를 기각할 기준선으로, 보통 0.05를 많이 쓴다.

개념설명
- α는 “이 정도까지는 우연으로 봐주겠다” 같은 기준이다.
- α=0.05면, H0가 참인데도 실수로 기각할 위험(1종 오류)을 5% 수준으로 제한하겠다는 뜻으로 이해하면 된다

## 의사결정 규칙(p-value vs α)
p ≤ α면 H0 기각, p > α면 H0를 기각하지 않는다.

개념설명
- p가 α보다 작다는 말은 “H0가 맞다면 이런 결과가 나올 확률이 기준선보다 더 작다”는 뜻이다. <br>
→ “우연으로 보기 어렵다” → H0를 버린다(기각).
- 대로 p가 크면 “우연일 수도 있겠다”<br>
→ H0를 버릴 만큼의 증거는 부족하다(기각 못함).

예시설명
- p=0.03, α=0.05 → 0.03 ≤ 0.05 → H0 기각
- p=0.30, α=0.05 → 0.30 > 0.05 → H0 기각 못함

## “기각 못함”의 정확한 의미(자주 틀리는 해석)
H0 기각 못함 = H0가 참이라는 증명이 아니라, 증거가 부족하다는 뜻이다.

개념설명
- p가 크면 “H0일 때도 충분히 가능한 결과”라는 뜻일 뿐이다.
- 그래서 “차이가 없다”를 확정할 수는 없다.
- 표본이 적으면 실제로 차이가 있어도 못 잡을 수 있다.

예시설명
- 표본이 너무 적은 A/B 테스트에서 p=0.2가 나왔다면 <br>
“개선이 없다”가 아니라 “개선이 있다고 말할 만큼 증거가 부족”일 수 있다.


예시로 확인하기

상황설정
- H0: 신약 효과 없음 (신약군−위약군 평균 차이 = 0)
- H1: 신약 효과 있음 (차이 ≠ 0)
- 실험 결과: 신약군 평균이 위약군보다 +5점 높게 나옴

가설 설정
- 만약 만약 진짜로 신약 효과가 0이라면(H0가 참이라면) 표본뽑기 운(우연)만으로도 +5점 같은 차이가 나올 확률이 얼마나 되냐?<br>→ 이 확률이 p-value다.

가설 증명
- p-value가 0.8인 경우
- 효과가 없다고 가정해도 우연만으로 지금 관측된 차이 정도는 아주 흔하게 나온다 <br>
→ 이 정도 차이는 우연이어도 충분히 나오는구나 → H0를 기각할 근거가 없다

- p-value가 0.01인 경우
- 효과가 없다고 가정했는데도, 우연만으로 이 정도(+5점) 또는 그보다 더 큰 차이가 나올 확률이 1%밖에 없다  <br> → “우연치고는 너무 희귀하다” → H0를 기각 신약은 효과가 있다(H1)

주의사항
- p-value가 작다고 해서 100% 무조건 기준 초과가 확정되는 건 아니다.
- p-value가 크다고 해서 100% 무조건 기준 이내가 확정되는 것도 아니다.
- 가설검정은 “확정”이 아니라 현재 데이터로 H0를 버릴 만큼 증거가 충분한지를 판단하는 절차다.
